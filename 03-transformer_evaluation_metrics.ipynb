```python
import pandas as pd
import os
from evaluate import load
```

```python
!wget -O 'transformer_evaluation_table.zip' $'https://www.dropbox.com/scl/fi/2249htzkm3f60yqwyc4rb/transformer_evaluation_table.zip?rlkey=r22cqlkd75hl6htsh4fgb701r&dl=0'
!unzip -q $'/content/transformer_evaluation_table.zip' -d $'/content'
os.remove("/content/transformer_evaluation_table.zip")
```

```python
eval_table = pd.read_csv('/content/transformer_evaluation_table.csv')
```

```python
eval_table
```

```python
bleu = load("bleu")
bleu_results = bleu.compute(predictions=eval_table["pred_prompt"], references=eval_table["true_prompt"])
print(bleu_results)
```

```python
meteor = load('meteor')
meteor_results = meteor.compute(predictions=eval_table["pred_prompt"], references=eval_table["true_prompt"])
print(meteor_results)
```

```python
bertscore = load("bertscore")
bert_results = bertscore.compute(predictions=eval_table["pred_prompt"], references=eval_table["true_prompt"], model_type="bert-large-uncased")
print(bert_results)
```

```python
eval_table['bert_precision'] = bert_results['precision']
eval_table['bert_recall'] = bert_results['recall']
eval_table['bert_f1'] = bert_results['f1']
```

```python
eval_table.bert_precision.mean(), eval_table.bert_recall.mean(), eval_table.bert_f1.mean()
```
